```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)

for i in range(9):
    # Load data
    amp_list_combined = np.load(f'data/amp_list_channel_{i+1}_500k.npy')
    # Sum along axis 1 and normalize
    summed = np.sum(amp_list_combined, axis=1) / 50
    # Define eta bins
    eta_list = np.linspace(-3, 3, 50)
    # Create a new figure for each channel
    plt.figure()
    # Plot histogram
    plt.hist(eta_list, bins=eta_list, weights=summed)
    plt.ylim(0, 1)
    # Add labels and title
    plt.xlabel(r"$\eta$")
    plt.ylabel(r"$\textrm{Weighted amplitude}$")
    plt.title(rf"$\textrm{{500k events}}$; ${500+250*i} \textrm{{GeV}} < m_{{jj}} < {750+250*i} \textrm{{GeV}}$")
    plt.savefig(f"plots/500k_rot/amp_hist_channel_{i+1}_500k.png")
# Show all figures
plt.show()
```
