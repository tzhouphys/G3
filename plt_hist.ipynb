```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
amp_list_combined = np.load(f'data/amp_list_channel_7.npy')
summed = np.sum(amp_list_combined, axis=1)/50
eta_list = np.linspace(-3, 3, 50)
plt.hist(bins=50, x=eta_list, weights=summed)
plt.xlabel(r"$\eta$")
plt.ylabel(r"$\textrm{Weighted amplitude}$")
plt.title(r"$\textrm{500k events}$; $2000\textrm{GeV} < m_{jj} < 2250 \textrm{GeV}$")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

for i in range(9):
    # Load data
    amp_list_combined = np.load(f'data/amp_list_channel_{i+1}.npy')
    # Sum along axis 1 and normalize
    summed = np.sum(amp_list_combined, axis=1) / 50
    # Define eta bins
    eta_list = np.linspace(-3, 3, 50)
    # Create a new figure for each channel
    plt.figure()
    # Plot histogram
    plt.hist(eta_list, bins=eta_list, weights=summed)
    plt.ylim(0, 1)
    # Add labels and title
    plt.xlabel(r"$\eta$")
    plt.ylabel(r"$\textrm{Weighted amplitude}$")
    plt.title(rf"$\textrm{{500k events}}$; $100 \textrm{{GeV}} < m_{{jj}} < 200 \textrm{{GeV}}$ (Channel {i+1})")
    # Uncomment the line below to save each figure
    # plt.savefig(f"amp_combined500k_hist_channel_{i+1}.png")
# Show all figures
plt.show()
```
