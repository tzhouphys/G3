```python
import numpy as np
import matplotlib.pyplot as plt 

amp_list_10k_100x100 = np.load('amp_list_pk_kk_combined.npy')
phi_list = np.linspace(0, 2*np.pi, 100)
eta_list = np.linspace(-3, 3, 100)
mesh_phi, mesh_eta = np.meshgrid(phi_list, eta_list)

plt.rc('text', usetex=True)
summed = np.sum(amp_list_10k_100x100, axis=1) / 100
plt.figure()
plt.hist(eta_list, bins=50, weights=summed)
#plt.ylim(0, 1)
plt.xlabel(r"$\eta$")
plt.ylabel(r"$\textrm{Weighted amplitude}$")
plt.title(rf"$\textrm{{10k events, 100x100}}$; $2000 \textrm{{GeV}} < m_{{jj}} < 2250 \textrm{{GeV}}$")
plt.show()
```
