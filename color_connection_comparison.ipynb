```python
import numpy as np
import matplotlib.pyplot as plt

amp_list_combined = np.load('amp_list_pk_kk_combined.npy')
#amp_list_pk = np.load('amp_list_pk.npy')
#amp_list_k1k2 = np.load('amp_list_k1k2.npy')
phi_list = np.linspace(0, 2*np.pi, 100)
eta_list = np.linspace(-3, 3, 100)
mesh_phi, mesh_eta = np.meshgrid(phi_list, eta_list)

plt.rc('text', usetex=True)
plt.xlabel(r"$\eta$")
plt.contourf(mesh_eta, mesh_phi, amp_list_combined, levels=np.linspace(0, 10, 10))
plt.colorbar()
plt.ylabel(r"$\phi$")
plt.title(r"Weighted amplitude averaged over color connections (500k events)")
#plt.savefig("amp_combined500k.png")
```
